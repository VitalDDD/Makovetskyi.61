```python
# Перше завдання
import cv2
import matplotlib.pyplot as plt

# Запит користувача для введення шляху до зображення
image_path = input("Введіть шлях до зображення: ")

# Завантаження зображення за вказаним шляхом
image = cv2.imread(image_path)

# Перетворення зображення у формат JPEG та задання коефіцієнта компресії
compression_quality = int(input("Введіть коефіцієнт компресії (від 0 до 100): "))

# Генерація нового імені файлу після компресії
output_filename = input("Введіть нову назву файлу після компресії: ")

# Застосування алгоритму стиснення JPEG
compression_params = [cv2.IMWRITE_JPEG_QUALITY, compression_quality]
cv2.imwrite(output_filename, image, compression_params)

# Завантаження стиснутого зображення
compressed_image = cv2.imread(output_filename)

# Відображення обох зображень поруч
fig, axs = plt.subplots(1, 2, figsize=(15, 10))
axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Оригінал')
axs[0].axis('off')
axs[1].imshow(cv2.cvtColor(compressed_image, cv2.COLOR_BGR2RGB))
axs[1].set_title('Стиснуто')
axs[1].axis('off')
plt.show()

```

```python
# Третє завдання

import heapq
import os

class HuffmanNode:
    def __init__(self, char, freq):
        self.char = char
        self.freq = freq
        self.left = None
        self.right = None

    def __lt__(self, other):
        return self.freq < other.freq


class HuffmanCompressor:
    def __init__(self):
        self.codes = {}

    def build_frequency_table(self, text):
        frequency_table = {}
        for char in text:
            if char not in frequency_table:
                frequency_table[char] = 0
            frequency_table[char] += 1
        return frequency_table

    def build_huffman_tree(self, frequency_table):
        heap = []
        for char, freq in frequency_table.items():
            heapq.heappush(heap, HuffmanNode(char, freq))

        while len(heap) > 1:
            node1 = heapq.heappop(heap)
            node2 = heapq.heappop(heap)
            merged = HuffmanNode(None, node1.freq + node2.freq)
            merged.left = node1
            merged.right = node2
            heapq.heappush(heap, merged)

        return heap[0]

    def build_codes_helper(self, root, current_code):
        if root is None:
            return

        if root.char is not None:
            self.codes[root.char] = current_code
            return

        self.build_codes_helper(root.left, current_code + "0")
        self.build_codes_helper(root.right, current_code + "1")

    def build_codes(self, root):
        self.build_codes_helper(root, "")

    def compress(self, text):
        frequency_table = self.build_frequency_table(text)
        huffman_tree = self.build_huffman_tree(frequency_table)
        self.build_codes(huffman_tree)

        encoded_text = ""
        for char in text:
            encoded_text += self.codes[char]

        padding = 8 - len(encoded_text) % 8
        encoded_text += padding * "0"
        padded_info = "{0:08b}".format(padding)
        encoded_text = padded_info + encoded_text

        bytes_array = bytearray()
        for i in range(0, len(encoded_text), 8):
            byte = encoded_text[i:i + 8]
            bytes_array.append(int(byte, 2))

        return bytes_array

    def decompress(self, encoded_bytes):
        binary_string = ""
        for byte in encoded_bytes:
            binary_string += "{0:08b}".format(byte)

        padding = int(binary_string[:8], 2)
        binary_string = binary_string[8:-padding]

        current_code = ""
        decompressed_text = ""
        for bit in binary_string:
            current_code += bit
            if current_code in self.codes:
                char = self.codes[current_code]
                decompressed_text += char
                current_code = ""

        return decompressed_text


# Запит шляху до файлу від користувача
input_file_path = input("Введіть шлях до вихідного файлу: ")
output_file_path = input("Введіть шлях до вихідного файлу після стиснення: ")

# Зчитування вмісту файлу
with open(input_file_path, 'r') as file:
    text = file.read()
# Стиснення тексту
compressor = HuffmanCompressor()
compressed_data = compressor.compress(text)

# Запис стисненого вмісту у новий файл
with open(output_file_path, 'wb') as file:
    file.write(compressed_data)

print("Файл був успішно стиснутий та збережений за вказаним шляхом:", output_file_path)

# Розпакування стисненого файлу
decompressed_data = compressor.decompress(compressed_data)

# Запис розпакованого вмісту у файл
decompressed_file_path = os.path.splitext(output_file_path)[0] + "_decompressed.txt"
with open(decompressed_file_path, 'w') as file:
    file.write(decompressed_data)

print("Файл був успішно розпакований та збережений за шляхом:", decompressed_file_path)
```
